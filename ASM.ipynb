```python
# ==========================
# main.py
# ==========================
import ASM

menu = {
    "1": "Nhập danh sách nhân viên",
    "2": "Đọc thông tin nhân viên từ file",
    "3": "Tính và hiển thị thông tin nhân viên theo mã",
    "4": "Xóa nhân viên theo mã, cập nhật file dữ liệu",
    "5": "Cập nhật thông tin nhân viên theo mã và cập nhật file dữ liệu",
    "6": "Tìm các nhân viên có lương trong khoảng nhập từ bàn phím",
    "7": "Sắp xếp nhân viên theo họ và tên",
    "8": "Sắp xếp nhân viên theo lương",
    "9": "Xuất 5 thông tin nhân viên có lương cao nhất",
    "0": "Thoát chương trình"
}

while True:
    print("\n========================")
    for k, v in menu.items():
        print(f"{k}. {v}")
    lua_chon = input("Nhập lựa chọn của bạn: ")

    if lua_chon == "0":
        print(">>> Kết thúc chương trình.")
        break
    elif lua_chon == "1":
        ASM.nhap_ds_nv()
    elif lua_chon == "2":
        ASM.doc_ds_nv()
    elif lua_chon == "3":
        ASM.tim_nv_ma()
    elif lua_chon == "4":
        ASM.xoa_nv()
    elif lua_chon == "5":
        ASM.capnhat_nv()
    elif lua_chon == "6":
        ASM.tim_nv_luong()
    elif lua_chon == "7":
        ASM.sapxep_ten()
    elif lua_chon == "8":
        ASM.sapxep_thunhap()
    elif lua_chon == "9":
        ASM.top5_nv()
    else:
        print(">>> Lựa chọn không hợp lệ, vui lòng nhập lại!")
```
